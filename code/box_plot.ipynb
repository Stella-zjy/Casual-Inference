```python
import matplotlib.pyplot as plt
import numpy as np

two_sls = [0.2] * 10
deepiv = [0.2] * 10
cevae = [0.164, 0.172, 0.142, 0.141, 0.152, 0.147, 0.157, 0.158, 0.158, 0.144]
data = [two_sls, deepiv, cevae]

# data = np.random.normal(100,20,200)
figure = plt.figure(figsize=(5,3))

ax = figure.add_axes([0,0,1,1])

bp = ax.boxplot(data)

ax.set_xticklabels(['2SLS', 'deepIV',
                    'CEVAE'])
 
# Adding title
plt.title("2SLS vs deepIV vs CEVAE on US Adult Cencus Income")

ax.get_xaxis().tick_bottom()
ax.get_yaxis().tick_left()

plt.show()
```
